# Preparing Training & Test Data On a Time Series Problem
For predictive maintenance problems, a time-dependent spliting strategy is often a better approach. <br>
A single point in time is chosen, the model is trained on examples up to that point in time, and validated on the examples after that point.<br> 
<span style="color:red">However, care must be taken on labels near the split point: registros anteriores al spliting point que se asocian con una falla NO pueden tener label de "próximo a falla" porque los datos futuros no se tienen --> hay que remediar esto en el dataset.</span>

```python
split_date = pd.to_datetime("2022-01-15")
fail_window = 2 * 24 # En horas
```

```python
# split the data into train/test based on date
df_train = df_input[df_input.timestamp < split_date]
df_test = df_input[df_input.timestamp >= split_date]

print(df_train.shape[0])
print(df_test.shape[0])
print('Porcentaje Testing:', round((df_test.shape[0]/df_train.shape[0])* 100, 2), '%')
```

```python
# Se analiza si se deben corregir los labels cercanos a la fecha de split. Los gráficos muestran que no hay fallas cerca
# de la fecha de split
# En caso de haberlas, se deberían volver a 0 los labels que estén dentro del período de ventana 1 y 2 (2y 4 días)
%matplotlib inline
lst_elements = pd.unique(df_input.element)
#df_plot = df_input[(df_input.timestamp.between(split_date+pd.Timedelta(hours=-(fail_window*2)), split_date))]\
df_plot = df_input[(df_input.timestamp.between('2021-09-01', '2022-02-20'))]\
            [['element', 'timestamp', 'about_to_fail_2d', 'about_to_fail_4d']]
fig, axs = plt.subplots(len(lst_elements), sharex=True, figsize=(15, 15), tight_layout=True)
fig.suptitle('Time Series dataset split')
fig.supxlabel('timestamp')
index=0
for element in lst_elements:
    axs[index].plot('timestamp', 'about_to_fail_2d', data=df_plot[df_plot.element==element], marker='.', linestyle='none')
    axs[index].plot('timestamp', 'about_to_fail_4d', data=df_plot[df_plot.element==element], marker='.', linestyle='none')
    axs[index].axvline(x=pd.to_datetime('2022-01-15'), c='r')
    axs[index].text(0.02, 0.1, element, horizontalalignment='center', verticalalignment='center', 
                    transform=axs[index].transAxes, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round'))
    axs[index].annotate('Set Split', xytext=(pd.to_datetime('2022-01-20'), 0.5), textcoords='data',
                        xy=(pd.to_datetime('2022-01-15'), 0),  xycoords='data',                        
                        arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),)
    axs[index].set_ylim(bottom=0, top=2.1)
    index+=1
for ax in fig.get_axes():
    ax.legend(loc=7) # 1=upper right, 7=center right
    ax.label_outer()
    # Show texts

plt.show()
```

# Imbalanced Dataset
In predictive maintenance, machine failures are usually rare occurrences compared to normal operation. This causes an imbalance in the label distribution and leads to poor performance as algorithms tend to classify majority class examples, since the total misclassification error is much improved when majority class is labeled correctly. <br>
This causes **low recall or precision rates**, although **accuracy can be high**. <br>
To help with this problem, sampling techniques such as oversampling of the minority examples can be used. <br>
